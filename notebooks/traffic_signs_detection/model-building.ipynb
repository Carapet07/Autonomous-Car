```python
from ultralytics import YOLO
from PIL import Image
import cv2
```

```python
# Use a pretrained YOLOv8 model
def yolo_model_building(epochs: int, batches: int):    
    model = YOLO('yolov8n.pt')

    hisory = model.train(data="../../Datasets/TrafficSignsSet/car/data.yaml",
                         epochs=epochs,
                         imgsz=416,
                         batch=batches, 
                         optimizer='auto',
                         project='../../models/traffic_signs_detection',
                         name='yolov8n_traffic_signs')
# Batch=-1 Auto batch size — let YOLO find the largest batch size your GPU can handle 
# without running out of VRAM
```
